# 이상치 탐색

```python
# box plot을 통한 이상치 탐색
cols = data.columns

f, axes = plt.subplots(10, 4)
f.set_size_inches((15, 30))
plt.subplots_adjust(wspace = 0.3, hspace = 0.3)

n = 0

for x in range(10):
  for y in range(4):
    n += 1
    if n == 40:
      break
    else:
      axes[x, y].boxplot(data[cols[n]])
      axes[x, y].set_title(cols[n])

plt.show()
```

이상치가 없는 column도, 이상치가 심한 column도 존재함

그러나 단순 EDA가 목적이므로 따로 처리하진 않을 예정

# 연속형 변수 분포 확인

```python
# 히스토그램을 통한 연속형 변수 분포 확인(binary 변수도 코드 편의상 포함하여 그림)
f, axes = plt.subplots(10, 4)
f.set_size_inches((15, 30))
plt.subplots_adjust(wspace = 0.3, hspace = 0.3)

n = 0

for x in range(10):
  for y in range(4):
    n += 1
    if n == 40:
      break
    else:
      axes[x, y].hist(data[cols[n]])
      axes[x, y].set_title(cols[n])

plt.show()
```

# Binary 변수 비율 확인

```python
# binary 변수 비율 확인
plt.bar(x=['0', '1'], height=[data.shape[0]-data['pctUrban'].sum(), data['pctUrban'].sum()])
plt.title('pctUrban')
plt.show()	

plt.bar(x=['0', '1'], height=[data.shape[0]-data['LemasPctOfficDrugUn'].sum(), data['LemasPctOfficDrugUn'].sum()])
plt.title('LemasPctOfficDrugUn')
plt.show()	
```

# 변수간 상관관계 확인

```python
# heatmap을 통한 변수별 상관관계 확인

plt.figure(figsize=(13, 10))

corr = data.corr()

sns.heatmap(corr, xticklabels=corr.columns, yticklabels=corr.columns)
```

칼럼이 너무 많아서 시각적으로 확인 어려움

시각화를 위해 naive한 변수 선택 진행(violperpop를 target 변수로 놓고, correlation 0.5 초과인 것들만 선택)

```python
data_selected = data.loc[:, corr['ViolentCrimesPerPop'][(abs(corr['ViolentCrimesPerPop'].sort_values())>0.5)].index]
```

```python
# heatmap을 통한 변수별 상관관계 확인

plt.figure(figsize=(13, 10))

corr2 = data_selected.corr()

sns.heatmap(corr2, xticklabels=corr2.columns, yticklabels=corr2.columns, annot=True)
```

기본적으로 모든 변수들 간의 상관관계가 높은 편. 

data description에서부터 애초에 범죄와 연관이 높다고 알려진 변수들만 선택된 data라고 명시된 만큼, data 특성상 서로 연관관계가 높을 수밖에 없다고 생각할 수 있음

몇 가지 눈에 띄는 점은, pctfam2par(두 부모가 가장인 가족(자녀가 있는)의 비율), pcthousownocc(가구주 점유율)는 범죄에 대해서 음의 상관 관계를 가지고 있는 것으로 보이며, 유일하게 이 둘만 다른 변수들에 대해서도 음의 상관 관계를 가지고 있는 것으로 보임




# 추가적인 패턴 확인

```python
# scatter plot을 통해 violentcrimesperpop(강력범죄수) vs nonviolperpop(비강력범죄수) 관계 비교
plt.scatter(data['ViolentCrimesPerPop'], data['nonViolPerPop'])
```

양의 상관 관계 -> 강력범죄+비강력범죄 합쳐서 범죄 변수로의 통합 필요성 제기

```python
# 인종과 강력범죄 수의 상관관계가 정말 인종별로 차이가 있을까?
f, axes = plt.subplots(1,3)
f.set_size_inches((20, 5))
plt.subplots_adjust(wspace = 0.3, hspace = 0.3)

axes[0].scatter(data['racepctblack'], data['ViolentCrimesPerPop'])
axes[0].set_title('racepctblack')

axes[1].scatter(data['racePctAsian'], data['ViolentCrimesPerPop'])
axes[1].set_title('racePctAsian')

axes[2].scatter(data['racePctHisp'], data['ViolentCrimesPerPop'])
axes[2].set_title('racePctHisp')

plt.show()
```

인종별 강력범죄 수와의 상관관계 비교

인종별로 차이가 있는지 비교해보았을때, 상대적으로 black과 hisp는 어느 정도의 양의 상관 관계를, asian은 비교적 상관 관계가 없음을 확인할 수 있음

(인종차별주의자 아닙니다....ㅠㅠㅠ)

```python
# 인종과 강력범죄 수의 상관관계가 정말 인종별로 차이가 있을까?
f, axes = plt.subplots(1,3)
f.set_size_inches((20, 5))
plt.subplots_adjust(wspace = 0.3, hspace = 0.3)

axes[0].scatter(data['racepctblack'], data['PctUnemployed'])
axes[0].set_title('racepctblack')

axes[1].scatter(data['racePctAsian'], data['PctUnemployed'])
axes[1].set_title('racePctAsian')

axes[2].scatter(data['racePctHisp'], data['PctUnemployed'])
axes[2].set_title('racePctHisp')

plt.show()
```

완벽한 상관관계라고 보기는 어렵지만, black과 hisp 비율이 높은 지역은 unemployed 비율도 유의미하게 높음! asian은 여기서도 상관관계가 크지 않고..

따라서 인종 비율에 따라 범죄율이 결정된다는 결론보다는 그 밑에 숨어 있는 실업률 문제로 해석하는 것이 보다 근본적인 원인일 수 있음
