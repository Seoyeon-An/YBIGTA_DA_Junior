# 1. Subword Tokenizer에 대해 알아보기

1) Byte Pair Encoding

Byte Pair Encoding(BPE)는 연속적으로 가장 많이 등장한 글자의 쌍을 찾아서 하나의 글자로 병합하는 방식을 더 이상 병합할 쌍이 없을 때까지 수행하는 방식을 의미합니다. 이러한 방식대로 subword tokenizer로서 BPE는 단어와 단어의 등장 횟수를 기록한 딕셔너리의 모든 단어들을 글자 단위로 분리하고, 이를 단어 집합으로 만듭니다. 이후 사용자가 정한 iteration에 따라 가장 빈도 수가 높은 유니그램의 쌍을 하나의 유니그램으로 통합하는 과정을 통해 최종적인 단어 집합을 만들어내는 방식으로 작동합니다.


2) SentencePiece

센텐스피스는 BPE 및 다른 서브워드 토크나이징 알고리즘들을 내장한 unsupervised text tokenizer 패키지입니다. 센텐스피스는 pre-tokenization 작업 없이 단어 분리 토큰화를 수행하므로, raw sentence로부터 training할 수 있으며, 모든 언어에 대해 적용이 가능하다는 특징이 있습니다. 또한 character-coverage 설정 및 subword generalization이 가능하며, 최적화가 잘 되어있어 속도가 빠르다는 장점이 있습니다.


3) Huggingface Tokenizer

Huggingface가 개발한 패키지인 tokenizers는 자주 등장하는 서브워드들을 하나의 토큰으로 취급하는 다양한 서브워드 토크나이저를 제공합니다. training과 tokenization 모두에서 빠른 속도를 보이며, truncation, padding, add_special_tokens 등과 같은 pre-processing에 필요한 것들을 대부분 지원하고 있습니다.
